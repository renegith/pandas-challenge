```python
#Player Count
pcount = len(set(purchase_df["SN"]))
```

```python
#Purchasing Analysis
#Number of unique items
uniqueitems = len(set(purchase_df["Item ID"]))
#Average purchase price
avgpp = "${:,.2f}".format(purchase_df["Price"].mean())
#Total number of purchases
totpur = len(purchase_df["Purchase ID"])
#Total revenue
totrev = "${:,.2f}".format(purchase_df["Price"].sum())
summary_pur_df = pd.DataFrame({"Number of Unique Items": uniqueitems, "Average Price": avgpp, "Number of Purchases": totpur, "Total Revenue": [totrev]})
summary_pur_df
```

```python
#Gender Demographics
purchase_df_m = purchase_df[purchase_df["Gender"] == "Male"]
purchase_df_f = purchase_df[purchase_df["Gender"] == "Female"]
purchase_df_o = purchase_df[purchase_df["Gender"] == "Other / Non-Disclosed"]
#Percentage and Count of Male Players
malecount = len(set(purchase_df_m["SN"]))
maleper = "{:.2%}".format(malecount/pcount)
#Percentage and Count of Female Players
femalecount = len(set(purchase_df_f["SN"]))
femaleper = "{:.2%}".format(femalecount/pcount)
#Percentage and Count of Other / Non-Disclosed
ocount = len(set(purchase_df_o["SN"]))
oper = "{:.2%}".format(ocount/pcount)
#SummaryTable
summary_demo_df = pd.DataFrame({"Gender": ["Male","Female","Other / Non-Disclosed"],
                                "Total Count":[malecount,femalecount,ocount],
                               "Percentage of Players":[maleper,femaleper,oper]})
summary_demo_df
```

```python
#Purchasing Analysis (Gender)
#The below each broken by gender
#Purchase Count
mpurcou = len(purchase_df_m["Purchase ID"])
fpurcou = len(purchase_df_f["Purchase ID"])
opurcou = len(purchase_df_o["Purchase ID"])
#Average Purchase Price
mavgpp = "${:,.2f}".format(purchase_df_m["Price"].mean())
favgpp = "${:,.2f}".format(purchase_df_f["Price"].mean())
oavgpp = "${:,.2f}".format(purchase_df_o["Price"].mean())
#Total Purchase Value
mtotrev = purchase_df_m["Price"].sum()
ftotrev = purchase_df_f["Price"].sum()
ototrev = purchase_df_o["Price"].sum()
#Average Purchase Total per Person by Gender
mavgtrev = mtotrev/malecount
favgtrev = ftotrev/femalecount
oavgtrev = ototrev/ocount
#SummaryTable
summarypagdf = pd.DataFrame({"Gender": ["Male","Female","Other / Non-Disclosed"],
                                "Purchase Count": [mpurcou,fpurcou,opurcou],
                               "Average Purchase Price":[mavgpp,favgpp,oavgpp],
                               "Total Purchase Value": ["${:,.2f}".format(mtotrev),"${:,.2f}".format(ftotrev),"${:,.2f}".format(ototrev)],
                               "Avg Total Purchase per Person": ["${:,.2f}".format(mavgtrev),"${:,.2f}".format(favgtrev),"${:,.2f}".format(oavgtrev)]})
summarypagdf
```
