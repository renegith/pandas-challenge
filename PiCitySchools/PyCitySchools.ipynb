```python
#1. Charter schools are much better than District because they have a much higher overall passing rate.
#2. District schools have problems teaching math, they usually have low passing rates.
#3. Despite of District schools usually having a higher budget per student, they are still not having good results, there is clearly a very big opportunity to imporve there.

# Dependencies and Setup
import pandas as pd
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#District Summary
#Calculate the total number of schools
totschools = len(set(school_data_complete["school_name"]))
#Calculate the total number of students
totstudents = len(set(school_data_complete["Student ID"]))
#Calculate the total budget
totbudget = "${:,.2f}".format(school_data["budget"].sum())
#Calculate the average math score 
avgmathscore = "{:.2%}".format((student_data["math_score"].mean()/100))
#Calculate the average reading score
avgreadscore = "{:.2%}".format((student_data["reading_score"].mean()/100))
#Calculate the percentage of students with a passing math score (70 or greater)
pms_df = student_data.loc[student_data["math_score"]>69]
pmsc = pms_df["Student ID"].count()
pmsr = "{:.2%}".format(pmsc/totstudents)
#Calculate the percentage of students with a passing reading score (70 or greater)
prs_df = student_data.loc[student_data["reading_score"]>69]
prsc = prs_df["Student ID"].count()
prsr = "{:.2%}".format(prsc/totstudents)
#Calculate the percentage of students who passed math and reading (% Overall Passing)
pt_df = student_data.loc[(student_data["math_score"] > 69) & (student_data["reading_score"] > 69)]
ptc = pt_df["Student ID"].count()
ptr = "{:.2%}".format(ptc/totstudents)
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Schools": totschools, "Total Students": totstudents, "Total Budget": totbudget,
                    "Average Math Score": [avgmathscore], "Average Reading Score": [avgreadscore], "% Passing Math": [pmsr],
                    "% Passing Reading": [prsr], "% Overall Passing": [ptr]})
district_summary.head()
#Optional: give the displayed data cleaner formatting
```

```python
#School Summary

#School Name
#School Type
schooltype = school_data.set_index(["school_name"])["type"]
schooltype

#Total Students per school
schoolstudents = school_data_complete["school_name"].value_counts()
schoolstudents
#Total School Budget
schoolbudget = school_data.groupby(["school_name"])["budget"].sum()
#Per Student Budget
studentbudget = schoolbudget / schoolstudents
#Average Math Score
amsperschool = round(school_data_complete.groupby(["school_name"])["math_score"].mean(),2)
amsperschool
#Average Reading Score
arsperschool = round(school_data_complete.groupby(["school_name"])["reading_score"].mean(),2)
arsperschool
#% Passing Math
mpass = school_data_complete[(school_data_complete["math_score"] > 69)]
mpass
pmsperschool = (mpass.groupby(["school_name"])["student_name"].count()/schoolstudents).map("{:.2%}".format)
pmsperschool
#% Passing Reading
rpass = school_data_complete[(school_data_complete["reading_score"] > 69)]
rpass
prsperschool = (rpass.groupby(["school_name"])["student_name"].count()/schoolstudents).map("{:.2%}".format)
prsperschool
#% Overall Passing (The percentage of students that passed math and reading.)
opass = school_data_complete.loc[(school_data_complete["math_score"] > 69) & (school_data_complete["reading_score"] > 69)]
opass
ppsperschool = (opass.groupby(["school_name"])["student_name"].count()/schoolstudents).map("{:.2%}".format)
ppsperschool          
#Create a dataframe to hold the above results

#Group by shoolname
school_summary = pd.DataFrame({"School Type": schooltype, "Total Students": schoolstudents, "Total School Budget": schoolbudget,
                    "Per Student Budget": studentbudget, "Average Math Score": amsperschool, "Average Reading Score": arsperschool,
                     "% Passing Math": pmsperschool, "% Passing Reading": prsperschool, "% Overall Passing": ppsperschool})
school_summary
```

```python
#Top Performing Schools (By % Overall Passing)
#Sort and display the top five performing schools by % overall passing.
topschools = school_summary.sort_values("% Overall Passing", ascending = False)
topschools
```

```python
#Bottom Performing Schools (By % Overall Passing)
#Sort and display the five worst-performing schools by % overall passing.
worstschools = school_summary.sort_values("% Overall Passing")
worstschools
```

```python
#Math Scores by Grade
#Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
grade9 = school_data_complete[(school_data_complete["grade"] == "9th")]
grade10 = school_data_complete[(school_data_complete["grade"] == "10th")]
grade11 = school_data_complete[(school_data_complete["grade"] == "11th")]
grade12 = school_data_complete[(school_data_complete["grade"] == "12th")]

m9 = (grade9.groupby(["school_name"]).mean()["math_score"]/100).map("{:.2%}".format)
m10 = (grade10.groupby(["school_name"]).mean()["math_score"]/100).map("{:.2%}".format)
m11 = (grade11.groupby(["school_name"]).mean()["math_score"]/100).map("{:.2%}".format)
m12 = (grade12.groupby(["school_name"]).mean()["math_score"]/100).map("{:.2%}".format)

# Groupby the above DataFrame by 'school_name' and calculate average on 'math_score' and then reset the index
mathscores = pd.DataFrame({"9th": m9,"10th": m10,"11th": m11,"12th": m12})
mathscores
```

```python
#Reading Scores by Grade
#Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
r9 = (grade9.groupby(["school_name"]).mean()["reading_score"]/100).map("{:.2%}".format)
r10 = (grade10.groupby(["school_name"]).mean()["reading_score"]/100).map("{:.2%}".format)
r11 = (grade11.groupby(["school_name"]).mean()["reading_score"]/100).map("{:.2%}".format)
r12 = (grade12.groupby(["school_name"]).mean()["reading_score"]/100).map("{:.2%}".format)

# Groupby the above DataFrame by 'school_name' and calculate average on 'reading_score' and then reset the index
readingscores = pd.DataFrame({"9th": r9,"10th": r10,"11th": r11,"12th": r12})
readingscores
```

```python
#Scores by School Spending
 
# creating spending bins
bins = [0, 585, 615, 645, 675]
group_name = ["<$585", "$585-615", "$615-645", "$645-675"]
school_data_complete["spending_bins"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], bins, labels = group_name)
#group by spending
byspending = school_data_complete.groupby("spending_bins")
#Average Math Score
avmbyschool = byspending["math_score"].mean()
#Average Reading Score
avrbyschool = byspending["reading_score"].mean()
#% Passing Math
pmbyschool = (school_data_complete[school_data_complete["math_score"] > 69].groupby("spending_bins")["Student ID"].count()/byspending["Student ID"].count()).map("{:.2%}".format)
#% Passing Reading
prbyschool = (school_data_complete[school_data_complete["reading_score"] > 69].groupby("spending_bins")["Student ID"].count()/byspending["Student ID"].count()).map("{:.2%}".format)
#Overall Passing Rate (Average of the above two)
pobyschool = (school_data_complete[(school_data_complete["math_score"] > 69) & (school_data_complete["reading_score"] > 69)].groupby("spending_bins")["Student ID"].count()/by_spending["Student ID"].count()).map("{:.2%}".format)
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
scoresbyspend = pd.DataFrame({"Average Math Score": avmbyschool,"Average Reading Score": avrbyschool,"% Passing Math": pmbyschool,"% Passing Reading": prbyschool,"Overall Passing Rate": pobyschool})
scoresbyspend
```

```python
#Scores by School Size
 
#Perform the same operations as above, based on school size.
 
# creating spending bins
bins = [0, 1000, 2000, 5000]
group_name = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete["size_bins"] = pd.cut(school_data_complete["size"], bins, labels = group_name)
#group by spending
bysize = school_data_complete.groupby("size_bins")
#Average Math Score
avmbysize = bysize["math_score"].mean()
#Average Reading Score
avrbysize = bysize["reading_score"].mean()
#% Passing Math
pmbysize = (school_data_complete[school_data_complete["math_score"] > 69].groupby("size_bins")["Student ID"].count()/bysize["Student ID"].count()).map("{:.2%}".format)
#% Passing Reading
prbysize = (school_data_complete[school_data_complete["reading_score"] > 69].groupby("size_bins")["Student ID"].count()/bysize["Student ID"].count()).map("{:.2%}".format)
#Overall Passing Rate (Average of the above two)
pobysize = (school_data_complete[(school_data_complete["math_score"] > 69) & (school_data_complete["reading_score"] > 69)].groupby("size_bins")["Student ID"].count()/bysize["Student ID"].count()).map("{:.2%}".format)
#Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
scoresbysize = pd.DataFrame({"Average Math Score": avmbysize,"Average Reading Score": avrbysize,"% Passing Math": pmbysize,"% Passing Reading": prbysize,"Overall Passing Rate": pobysize})
scoresbysize
```
